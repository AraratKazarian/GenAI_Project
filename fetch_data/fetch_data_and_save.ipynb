```python
URL = "https://seller-analytics-api.wildberries.ru/api/v2/nm-report/grouped/history"
```

```python
headers = {
    "Authorization": API_KEY,
    "Content-Type": "application/json"
}
```

```python
# Set date range
start_date = "2025-05-05"
end_date = datetime.now().strftime("%Y-%m-%d")

payload = {
    "period": {
        "begin": start_date,
        "end": end_date
    }
}
```

```python
# Fetch data and save to CSV
try:
    response = requests.post(URL, headers=headers, json=payload)
    response.raise_for_status()
    data = response.json()

    stats = data.get("data", [])
    if stats:
        df2 = pd.DataFrame(pd.DataFrame(stats).iloc[0].history)
        df2.to_csv("wildberries_data.csv", index=False)
        print(f"Saved {len(df2)} rows to 'wildberries_data.csv'.")
    else:
        print("No data returned for the selected period.")

except requests.exceptions.RequestException as e:
    print(f"Request failed: {e}")
    print(response.text)
```
